```python
import pandas as pd
import requests
import json
import ast

# Carregando o DataFrame
data = pd.read_csv('newly_fetched_data.csv')

# Definindo a chave da API e outros par√¢metros necess√°rios
API_KEY = "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"  # Substitua "XXXXXX" pela sua chave real
link = "https://api.openai.com/v1/chat/completions"
headers = {"Authorization": f"Bearer {API_KEY}", "Content-Type":"application/json"}
id_modelo = "gpt-3.5-turbo-0301"

# Inicializando o prompt
base_prompt = """
Sua √∫nica tarefa √© fornecer uma lista python dada a descri√ß√£o da vaga a seguir, forne√ßa uma lista Python com os seguintes elementos:

- O primeiro elemento √© um n√∫mero:
   1 para Analista de Dados (incluindo fun√ß√µes semelhantes como Analista de BI),
   2 para Cientista de Dados (incluindo Engenheiro de Machine Learning e Engenheiro de MLOPS),
   3 para Engenheiro de Dados,
   0 se for outro cargo ou n√£o identificado.

- O segundo elemento √© um n√∫mero:
   1 se a vaga for remota,
   0 caso contr√°rio.

- O terceiro elemento √© uma string contendo um resumo da vaga em tr√™s linhas, como se estivesse recomendando a vaga a um candidato.

- O quarto elemento √© uma string representando a senioridade da vaga: S√™nior, Pleno ou J√∫nior, CASO N√ÉO ESTEJA EXPLICITO, VOCE DEVE JULGAR BASEADO NA DESCRI√á√ÉO.

IMPORTANTE: A RESPOSTA DEVE SER apenas uma lista com esses quatro elementos, sem nenhum texto adicional.

Descri√ß√£o da vaga:
"""
final_warning = """
Lembre-se de formatar sua resposta da seguinte forma:
[ C√≥digo do Cargo, Trabalho Remoto, "Resumo da Vaga", "N√≠vel de Senioridade"]
"""
# Lista para armazenar os resultados
results = []

def fetch_openai_response(prompt):
    body_mensagem = {
        "model": id_modelo,
        "messages": [{"role": "user", "content": prompt}]
    }
    print(f"Enviando solicita√ß√£o para a API com o prompt:\n{prompt[:100]}...")  # Mostra os primeiros 100 caracteres do prompt
    requisicao = requests.post(link, headers=headers, data=json.dumps(body_mensagem))
    resposta = requisicao.json()
    # Verificando se "choices" est√° na resposta
    if "choices" in resposta:
        print("Resposta recebida da API:")
        print(resposta["choices"][0]["message"]["content"])  # Print do resultado da requisi√ß√£o
        return resposta["choices"][0]["message"]["content"]
    else:
        print("Erro! N√£o houve uma resposta v√°lida da API.")
        return None

# Iterando sobre as descri√ß√µes das vagas no DataFrame
for idx, description in enumerate(data["DESCRICAO"]):
    print(f"\nProcessando descri√ß√£o da vaga {idx+1} de {len(data['DESCRICAO'])}.")
    full_prompt = base_prompt + "\n" + description

    max_attempts = 3
    is_list = False
    attempts = 0

    while not is_list and attempts < max_attempts:
        mensagem = fetch_openai_response(full_prompt)
        if mensagem:
            try:
                lista = ast.literal_eval(mensagem)
                if isinstance(lista, list):
                    print(f"Sucesso ao processar a descri√ß√£o da vaga {idx+1}.")
                    is_list = True
                else:
                    attempts += 1
                    print(f"Tentativa {attempts} falhou. A resposta n√£o estava no formato de lista esperado.")
            except:
                attempts += 1
                print(f"Erro ao tentar converter a resposta em lista na tentativa {attempts}.")
        else:
            attempts += 1
            print(f"Tentativa {attempts} falhou. N√£o houve resposta da API.")

    if not is_list:
        lista = ["Erro", "Erro", "Erro", "Erro"]
        print(f"N√£o foi poss√≠vel processar a descri√ß√£o da vaga {idx+1} ap√≥s {max_attempts} tentativas.")

    results.append(lista)

print("\nProcessamento completo. Convertendo resultados em DataFrame...")
# Convertendo os resultados em um DataFrame
results_df = pd.DataFrame(results, columns=["Cargo", "Remoto", "Resumo", "Senioridade"])

# Juntando os DataFrames
final_df = pd.concat([data, results_df], axis=1)
print("DataFrame final criado com sucesso!")

```
